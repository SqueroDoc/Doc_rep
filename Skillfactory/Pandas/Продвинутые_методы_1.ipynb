```python
# Прочитаем датасет и выведем первые пять его строк:

import pandas as pd

melb_df = pd.read_csv('data/melb_data_fe.csv')
melb_df.head()
```

```python
# Файл подгрузился верно, однако есть один нюанс. Давайте посмотрим на информацию о столбцах с помощью метода info():

display(melb_df.info())
```

Обратите внимание, что наше преобразование столбцов к типам datetime и category «слетело».

?
Почему это произошло?

Ответ на самом деле очень прост: csv-файл не хранит в себе информацию о типах данных столбцов, поэтому при чтении Pandas автоматически определяет тип данных столбца. Не забывайте об этом, обмениваясь преобразованными данными с вашими коллегами.

```python
# Преобразуйте столбец Date в формат datetime и выделите квартал (quarter) продажи объектов недвижимости. Найдите второй по популярности квартал продажи. В качестве ответа запишите число объектов, проданных в этом квартале.

# Шаг 1: Преобразуем столбец Date в формат datetime
melb_df['Date'] = pd.to_datetime(melb_df['Date'])

# Шаг 2: Извлекаем квартал из столбца Date
melb_df['quarter'] = melb_df['Date'].dt.quarter

# Шаг 3: Подсчитываем количество объектов, проданных в каждом квартале
quarter_sales = melb_df['quarter'].value_counts()

# Шаг 4: Сортируем по количеству продаж и находим второй по популярности квартал
sorted_quarter_sales = quarter_sales.sort_values(ascending=False)

# Шаг 5: Получаем количество объектов, проданных во втором по популярности квартале
second_most_popular_quarter = sorted_quarter_sales.iloc[1]

# Вывод результата
print(f"Количество объектов, проданных во втором по популярности квартале: {second_most_popular_quarter}")
```

```python
# Преобразуйте все столбцы, в которых меньше 150 уникальных значений, в тип данных category, исключив из преобразования столбцы Date, Rooms, Bedroom, Bathroom, Car. В качестве ответа запишите результирующее количество столбцов, которые имеют тип данных category.


# Шаг 1: Определим список столбцов, которые не должны быть преобразованы
exclude_columns = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car']

# Шаг 2: Преобразуем столбцы в тип category, если количество уникальных значений меньше 150
for column in melb_df.columns:
    if column not in exclude_columns:  # Исключаем столбцы, которые не должны быть преобразованы
        unique_values_count = melb_df[column].nunique()
        if unique_values_count < 150:  # Преобразуем, если уникальных значений меньше 150
            melb_df[column] = melb_df[column].astype('category')

# Шаг 3: Подсчитаем количество столбцов с типом category
category_columns_count = melb_df.select_dtypes(include=['category']).shape[1]

# Вывод результата
print(f"Количество столбцов с типом данных 'category': {category_columns_count}")
```
