# Способы графического изображения экстенсивных показателей

Графически экстенсивный показатель можно изобразить только в виде секторной (круговой) или внутристолбиковой диаграммы. Рассмотрим правила построения таких диаграмм на примере с инфекционными заболеваниями из предыдущего урока.

Алгоритм построения секторной диаграммы:

1. Определение данных

Начните с определения данных, которые вы хотите представить на диаграмме. Это может быть распределение чего-либо по категориям или доли чего-то в общем объеме.

2. Расчет процентных долей

Для каждой категории вычислите ее процентное соотношение относительно общего объема данных. Проценты должны быть представлены в виде долей, равных дроби от 1. Например, 0.25, 0.5, 0.75.

3. Выбор порядка категорий

Установите порядок категорий, который будет отображен на диаграмме. Обычно порядок идет по часовой стрелке, от самого большого сектора к самому маленькому.

4. Рассчитайте углы секторов

Для каждой категории вычислите угол сектора, который соответствует ее процентной доле от общего угла круга (360 градусов). Угол можно рассчитать, умножив процент на 360.

5. Нарисуйте диаграмму

Используя углы секторов и порядок категорий, нарисуйте круговую диаграмму с соответствующими секторами для каждой категории. Убедитесь, что каждый сектор четко отражает свою процентную долю.

6. Добавьте легенду

Для большей понятности добавьте легенду, которая поясняет категории или данные, представленные на диаграмме.

7. Настройки

Не забудьте добавить заголовок и подписи осей, если необходимо. А также провести окончательное оформление диаграммы, чтобы сделать ее эстетически привлекательной.

Алгоритм построения внутристолбиковой диаграммы:

1. Определение данных

2. Выбор основного столбца

Определите основной столбец, на основе которого будет строиться внутристолбиковая диаграмма. Этот столбец будет выражать общий объем или сумму данных, с которыми будут сравниваться другие показатели.

3. Разделение столбца на части

Разделите основной столбец на части, равные долям или процентам каждого из показателей, которые вы хотите отобразить. Каждая часть будет соответствовать одному из сравниваемых показателей.

4. Отображение данных

Для каждого показателя отметьте его долю внутри основного столбца, используя различные элементы (например, цветовые сегменты или штриховку). При этом важно учесть масштабы и пропорции для четкого представления данных.

5. Легенда и объяснения

Добавьте легенду, которая пояснит, какому показателю соответствует каждая часть внутри столбца. Объясните основные выводы и сделанные сравнения.

6. Оформление и анализ

После построения внутристолбиковой диаграммы обратите внимание на оформление (например, добавление заголовков, шкал и подписей) для улучшения читаемости. Проанализируйте данные, сделайте выводы и подготовьте дополнительные иллюстрации или графики для их подтверждения.
