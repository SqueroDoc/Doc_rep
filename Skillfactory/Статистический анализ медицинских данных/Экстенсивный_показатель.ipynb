# Экстенсивный показатель

***Экстенсивный показатель*** — относительная величина, показывающая, как велика отдельная часть по отношению ко всей изучаемой совокупности. Одним из основных экстенсивных показателей, используемых в медицинских исследованиях, является общее количество заболевших в определенной популяции или общее количество случаев заболевания в определенном временном периоде.

Экстенсивные показатели могут быть также связаны с оценкой распространенности определенных заболеваний или сберегающих здоровье практик. Например, показатели, отражающие количество диагностированных случаев редких заболеваний в определенной популяции, помогут понять необходимость создания специализированных программ по их обнаружению и лечению.

Экстенсивные показатели также могут быть использованы для анализа эффективности профилактических программ или лечебных методов. На основе данных о количестве проведенных процедур, выписанных лекарств или успешно пройденных курсов реабилитации можно делать выводы о том, насколько эффективны эти меры для улучшения состояния пациентов и снижения заболеваемости.

Таким образом, экстенсивные показатели в медицине играют значительную роль в оценке здоровья населения, планировании медицинских программ и формировании стратегий по улучшению общественного здоровья.

Классификация экстенсивных показателей:
* Показатели удельного веса части в целом. Например, доля заболевших короновирусом среди всех заболеваний.
* Показатели структуры или распределения. Распределение всех заболеваний, зафиксированных за год, по отдельным заболеваниям.

Чтобы рассчитать экстенсивные показатели, нужно знать численность всей совокупности и составляющих ее частей (или отдельной части данной совокупности).

Чаще всего экстенсивный показатель рассчитывается в процентах. Вся совокупность принимается за 100%, отдельные ее части ― за «x». С помощью пропорции вычисляется процент доли от целого:

<math xmlns="http://www.w3.org/1998/Math/MathML" display="block"> <mtext> экстенсивный показатель </mtext> <mo>=</mo> <mfrac> <mtext> часть явления </mtext> <mtext> явление в целом </mtext> </mfrac> <mo>×</mo> <mn>100</mn> <mi mathvariant="normal">%</mi> </math>

Этот подход вычисляет, сколько процентов приходится на конкретную часть совокупности.

Расчет экстенсивных показателей:

1. Процент заболеваемости в определенной популяции

Для определения процента заболеваемости в популяции можно рассчитать отношение числа заболевших к общему числу людей в этой популяции и умножить на 100%.

Например, если в популяции из 1000 человек было 50 заболевших, то процент заболеваемости составит (50 / 1000) * 100% = 5%.

2. Общее количество случаев заболеваний

Для определения экстенсивного показателя общего количества случаев заболеваний в популяции за определенный период времени нужно просто сложить все единичные значения количества заболевших за каждый месяц или год.

Например, если у нас есть данные о числе заболевших в популяции каждый месяц за год, то, суммируя все значения, мы получим общее количество случаев заболеваний за весь год.

3. Суммарное количество проведенных операций

Для оценки экстенсивного показателя суммарного количества проведенных операций в больнице за год можно сложить все значения операций за каждый месяц.

Например, если в январе было проведено 100 операций, в феврале 120 операций и так далее, то общее количество операций за год будет равно сумме этих значений.

##### Пример
В Московской области было зарегистрировано 110 случаев инфекционных заболеваний за год. Краснухой заболели 50 человек, корью ― 35, паротитом ― 25. Нужно определить структуру инфекционных заболеваний и проанализировать полученные результаты.

Вся совокупность, то есть все 110 случаев заболеваний, принимаются за 100%. Удельный вес случаев заболевания краснухой составит:

<math xmlns="http://www.w3.org/1998/Math/MathML" display="block"> <mfrac> <mn>50</mn> <mn>110</mn> </mfrac> <mo>⋅</mo> <mn>100</mn> <mi mathvariant="normal">%</mi> <mo>≈</mo> <mn>45</mn> <mo>,</mo> <mn>5</mn> <mi mathvariant="normal">%</mi> </math>

Аналогично рассчитывается удельный вес кори и паротита.

Вывод: в структуре инфекционных заболеваний доля краснухи составила 45,5%, кори ― 31,8%, паротита ― 22,7%.
