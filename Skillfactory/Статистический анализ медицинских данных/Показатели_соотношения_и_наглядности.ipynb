# Показатели соотношения и наглядности

### Показатель соотношения
Характеризует соотношение между двумя совокупностями, не связанными между собой (соотношение абортов и родов, обеспеченность населения койками, врачами, соотношение врачей и другого медицинского персонала).

Чтобы вычислить показатель соотношения, нужны две совокупности.

Абсолютная величина, характеризующая совокупность №1, делится на абсолютную величину, характеризующую не связанную с ней совокупность №2, и умножается на множитель (100, 1000, 10 000 и так далее):

<math xmlns="http://www.w3.org/1998/Math/MathML" display="block"> <mtext> показатель соотношения </mtext> <mo>=</mo> <mfrac> <mrow> <mtext> совокупность </mtext> <mn>1</mn> </mrow> <mrow> <mtext> совокупность </mtext> <mn>2</mn> </mrow> </mfrac> <mo>×</mo> <mtext> масштабирующий коэффициент </mtext> </math>

##### Пример
Население города А ― 50 000 человек. Количество терапевтических коек в местной городской больнице ― 400. Требуется рассчитать обеспеченность населения города А терапевтическими койками.

Число коек ― это совокупность №1, численность населения ― совокупность №2.

<math xmlns="http://www.w3.org/1998/Math/MathML" display="block"> <mtext> обеспеченность койками </mtext> <mo>=</mo> <mfrac> <mn>400</mn> <mn>50000</mn> </mfrac> <mo>×</mo> <mn>10000</mn> <mo>=</mo> <mn>80</mn> <mtext> на </mtext> <mn>10000</mn> <mtext> человек </mtext> </math>

То есть в городе А на 10 000 населения приходится 80 терапевтических коек.

Графически показатель соотношения можно представить теми же диаграммами, что и интенсивный показатель: линейной, столбчатой, ленточной, радиальной.

### Показатель наглядности
Используется в анализе однородных чисел и применяется, когда необходимо «уйти» от показа истинных величин (относительных и средних величин, абсолютных чисел). Например, когда проводится сравнительный анализ одних и тех же показателей, но на разных территориях или в разное время. Часто эти величины представлены в динамике.

Чтобы вычислить показатель наглядности, одну из сравниваемых величин принимают за 100% (часто исходную величину). Остальные вычисляются в процентном отношении к ней.

### Итоги
Все виды относительных величин (интенсивные, экстенсивные, соотношения и наглядные) широко применяются в медицине для диагностики, контроля состояния здоровья и определения эффективности лечения различных заболеваний.

Использование как абсолютных, так и относительных величин в медицинской описательной статистике позволяет более полно и точно описывать и анализировать данные, что важно для принятия обоснованных медицинских решений и разработки эффективных методов лечения и профилактики различных заболеваний.
